```python
def amat2lmat(A):
    '''Generate Laplacian matrix from Adjacency matrix
'''
    B = (A + A.T)/2.
    D = np.sum(B,axis = -1) - np.diag(B)
    np.fill_diagonal(B,-D)
    return -B
```

```python
import tutil
X = tutil.toyData(K=4,D = 4,N0=150)
# D = 
D = np.sqrt(np.square(X[None] - X[:,None]).sum(axis=(-1)))
```

```python
# D = np.square(X  - X.T)
# c = 20
c = 4.
A = np.exp(-D/c)
L = amat2lmat(A)

pyvis.heatmap(A)
```
