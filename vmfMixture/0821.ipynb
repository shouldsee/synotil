Because the shape of the equi-potential surfaces are invariant/isomorphic for a particular multi-variable gaussian, it is possible to separate the distribution of potential 

$$
E = x^T \Sigma ^ {- 1} x
$$ 

from the distribution of the shape 

$$
x\in \{x | x^T \Sigma ^ {- 1} x = 1\}
$$

Traditionally, one assumes a uniform distribution for shape but an exponentional distribution for the potential.  However, in RNA-Seq data we have noticed that a Poisson distribution better capture the essence of count data, with PDF taking form

$$\begin{aligned}
\text{Pois}(X|\mu) &= \frac{ \mu ^ X \exp(- \mu) }{X!} 
\\ 
&=\frac{1 }{\Gamma(Y)}   \mu ^ {Y-1} \exp(- \mu) \,, Y = X + 1
\end{aligned}$$

(Not quite) Extrapolating this function to the continuous domain results in a gamma distribution.

Gamma distribution is known to be the conjugate prior for parameter $\mu$.

$$\begin{align}
\text{Gamma}(X|\, k, \theta) &= \frac{1}{\Gamma ( k) } \frac{1}{\theta ^ k} X^{k-1} \exp (-X /\theta)
\\
\text{Gamma}(X|\, k, 1/\beta) &= \frac{1}{\Gamma ( k) } \beta^k X^{k-1} \exp (-\beta X )
\\
\text{Gamma}(X|\, k, 1) &= \frac{1}{\Gamma ( k) }  X^{k-1} \exp (- X )
\end{align}$$

```python
def random_radVec(N=1,D = 2):
    C=  np.eye(D)
    X = np.random.multivariate_normal((0,)*D,C,size = N)
    Xn = X 
    Xn = Xn/np.linalg.norm(Xn,axis=1,keepdims=1)
    return Xn
```

* If ''X'' ~ Gamma(/2, 2)(shape -scale parametrization), then ''X'' is identical to ''''<sup>2</sup>(''''), the [[chi-squared distribution]] with '''' degrees of freedom. Conversely, if ''Q'' ~ ''''<sup>2</sup>('''') and ''c'' is a positive constant, then ''cQ'' ~ Gamma(''''/2, 2''c'').

* If ''k'' is an [[integer]], the gamma distribution is an [[Erlang distribution]] and is the probability distribution of the waiting time until the ''k''th "arrival" in a one-dimensional [[Poisson process]] with intensity 1/''''. If
::$X \sim \Gamma(k \in \mathbf{Z}, \theta) \\
\qquad Y \sim \mathrm{Pois}\left(\frac x \theta \right),
$
then 
$
P(X > x) = P(Y < k).$
